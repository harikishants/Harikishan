```python
import json

def extract_from_json(profile_json,activity_json):
    profile = json.load(open(profile_json))
    activity = json.load(open(activity_json))
    i=0
    person_data = {'fullname': profile[i]['general']['fullName'],
                  'linkedin profile': profile[i]['details']['linkedinProfile'],
                  'mail id': profile[i]['details']['mail'],
                  'connectionsCount':profile[i]['general']['connectionsCount'],
                  'description':profile[i]['general']['description'],
                  'jobsCount':len(profile[i]['jobs']),
                  'schoolsCount':len(profile[i]['schools']),
                  'posts':[]}

    for j in range(len(activity)):
        person_data['posts'].append({"content":activity[j]['postContent'],
               "likes":activity[j]['likeCount'],
               "comments":activity[j]['commentCount'],
               "views":activity[j]['viewCount']})
    return person_data

person = extract_from_json('profile.json','activity.json')
person
```

```python
import openai
import os

openai.api_key  = '#### USE YOUR API KEY ####'

def get_completion(messages, model="gpt-3.5-turbo"):
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        max_tokens=46, 
        #n=1, 
        #stop=None,
        temperature=0, # this is the degree of randomness of the model's output
    )
    #print(response.usage)
    return response.choices[0].message["content"]
```

```python
def scale_score(score):
    if score <= 35.0:
        return 1
    if score <= 70.0:
        return 2
    else:
        return 3
```

```python
def get_texts_score(texts):
    delimiter = '####'
    prompt = f"""
    Identify the following items and give rating from 0.0 to 100.0 from the text chunk delimitted by {delimiter}: 
    - Authenticity
    - Clout
    - Analytical thinking
    - Emotional tone

    The text chunk is delimited with {delimiter}
    Format your output response as a JSON object with keys as
    "authenticity": "11.5",
    "clout": "81.6",
    "analyticalThinking": "52.1",
    "emotionalTone": "80.2"
    DO NOT include anything in output other than the format specified

    If the information isn't present, use "0.0" as the value.
    Make your response is only floating point with TWO decimals ranging from 0.0 to 100.0
    Take your time to analyze the text properly without skipping anything.

    Text chunk: {delimiter}{texts}{delimiter}
    """

    messages =  [  
    {'role':'system', 'content':'You are an assistant that give sentiment rating to the given texts on four categories:\
    - Authenticity\
    - Clout\
    - Analytical thinking\
    - Emotional tone\
    '},     
    {'role':'user', 'content':prompt} ]
    
    reply = get_completion(messages)
    #print(reply)
    
    response = ast.literal_eval(reply)
    #print(float(response['clout'])+float(response['authenticity'])+0.12)
    score=0
    #print(float(response['clout']))
    score = scale_score(float(response['clout']))\
            +scale_score(float(response['authenticity']))\
            +scale_score(float(response['analyticalThinking']))\
            +scale_score(float(response['emotionalTone']))
    #print(score)
    return score
    
```

```python
def post_score_GPT(person):
    avg = 0
    for i in range(len(person['posts'])):
        #print(person['posts'][i]['content'])
        print('GPT processing Post {}'.format(i))
        avg += get_texts_score(person['posts'][i]['content'])
        time.sleep(20)
    print('Post_score = ',avg/len(person['posts']))
    return avg/len(person['posts'])
```

```python
def like_post_ratio(person):
    count=0
    for i in range(len(person['posts'])):
        count += person['posts'][i]['likes']
    
    if(count/len(person['posts'])<=10):
        print('like_post_ratio = 1')
        return 1
    if(count/len(person['posts'])<=20):
        print('like_post_ratio = 2')
        return 2
    else:
        print('like_post_ratio = 3')
        return 3
```

```python
def comment_post_ratio(person):
    count=0
    for i in range(len(person['posts'])):
        count += person['posts'][i]['comments']
    
    if(count/len(person['posts'])<=2):
        print('comment_post_ratio = 1')
        return 1
    if(count/len(person['posts'])<=5):
        print('comment_post_ratio = 2')
        return 2
    else:
        print('comment_post_ratio = 3')
        return 3
print(comment_post_ratio(person))
```

```python
def num_post_score(person):
    if len(person['posts'])<=10:
        print('num_post_score = 1')
        return 1
    else:
        print('num_post_score = 2')
        return 2
print(num_post_score(person))
```

```python
def conversational_score(person):
    score = 0
    score = post_score_GPT(person)\
            +like_post_ratio(person)\
            +comment_post_ratio(person)\
            +num_post_score(person)
    return score
print('Conversational score = ',conversational_score(person))
```

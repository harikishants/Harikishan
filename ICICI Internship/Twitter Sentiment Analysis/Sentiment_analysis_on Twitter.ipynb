```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.text import Tokenizer, text_to_word_sequence
from tensorflow.keras.preprocessing.sequence import pad_sequences

import bz2
from sklearn.metrics import f1_score, roc_auc_score, accuracy_score

%matplotlib inline
```

```python
train_texts = []
with open("train_texts.txt", "r") as f:
  for line in f:
    train_texts.append(line.strip())
```

```python
num_of_words = 3000 # Obtained while model building
tokenizer = Tokenizer(num_words=num_of_words)
tokenizer.fit_on_texts(train_texts)
max_length_of_sequences = 235 # Obtained while model building
```

```python
new_model_NN = tf.keras.models.load_model('Model_NN_100000.h5')
new_model_GRU = tf.keras.models.load_model('Model_GRU_100000.h5')
```

```python
test_text = "This is not normal, this is great"
mytest = tokenizer.texts_to_sequences([test_text])
mytest = pad_sequences(mytest, maxlen=max_length_of_sequences, padding='pre')
print("Output NN", new_model_NN.predict(mytest))
print("Output GRU", new_model_GRU.predict(mytest))
```

```python
import snscrape.modules.twitter as sntwitter
import pandas as pd

# Created a list to append all tweet attributes(data)
attributes_container = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('from:TimesNow').get_items()):
    if i>10:
        break
    attributes_container.append([tweet.date, tweet.likeCount, tweet.sourceLabel, tweet.renderedContent])
    
# Creating a dataframe from the tweets list above 
tweets_df = pd.DataFrame(attributes_container, columns=["Date Created", "Number of Likes", "Source of Tweet", "Tweets"])
```

```python
print(len(attributes_container),'\n')
for i in range(len(attributes_container)):
    test_text = attributes_container[i][3]
    mytest = tokenizer.texts_to_sequences([test_text])
    mytest = pad_sequences(mytest, maxlen=max_length_of_sequences, padding='pre')
    print("*****************START******************")
    print("Output NN", new_model_NN.predict(mytest))
    print("Output GRU", new_model_GRU.predict(mytest))
    
    print("Tweet:", attributes_container[i][3])
    print('******************END******************** \n\n')
```
